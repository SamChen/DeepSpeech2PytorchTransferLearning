```python
def counting_utts(spkid, dataset):
    assert isinstance(dataset, pd.DataFrame)
    count = 0
    for i in dataset.uttid.tolist():
        if spkid in i:
            count += 1
    return count

def counting_utts_for_allset(spkid):
    train_data = pd.read_csv("iconect/models/pt_only_all_utt/train.csv")
    val_data   = pd.read_csv("iconect/models/pt_only_all_utt/val.csv")
    test_data  = pd.read_csv("iconect/models/pt_only_all_utt/test.csv")
    
    print("train set, {}: {}".format(spkid, counting_utts(spkid, train_data)))
    print("val set, {}: {}".format(spkid, counting_utts(spkid, val_data)))
    print("test set, {}: {}".format(spkid, counting_utts(spkid, test_data)))
```

```python
uttids = train_data.uttid.tolist()
spkids = []
for i in uttids:
    spkids.append(i.split('_')[0])
spkids = list(set(spkids))
for spkid in spkids:
    print("=======================================")
    counting_utts_for_allset(spkid)
```
